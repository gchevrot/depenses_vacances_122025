```python
alimentation_avec_loulou = depenses_alimentation.alimentation_avec_loulou.sum()
print(f"Dépenses alimentation avec Loulou : {alimentation_avec_loulou} €")
```

```python
alimentation_sans_loulou = depenses_alimentation.alimentation_sans_loulou.sum()
print(f"Dépenses alimentation sans Loulou : {alimentation_sans_loulou} €")
```

## Repartition des dépenses alimentaires

### Avec Loulou

```python
total_adultes = 7
total_enfants = 7  # Rose n'est pas comptabilisée
coef_adulte = 1
coef_enfant = 1
coef_olivet = 2 * coef_enfant + 2 * coef_adulte
coef_crans = 1 * coef_enfant + 2 * coef_adulte
coef_pau = 2 * coef_enfant + 1 * coef_adulte
coef_chalon = 2 * coef_enfant + 2 * coef_adulte
```

```python
assert coef_olivet + coef_crans + coef_pau + coef_chalon == (total_enfants * coef_enfant + total_adultes * coef_adulte)
```

```python
alimentation_chalon = alimentation_avec_loulou * coef_chalon / (total_enfants * coef_enfant + total_adultes * coef_adulte)
alimentation_olivet_1 = alimentation_avec_loulou * coef_olivet / (total_enfants * coef_enfant + total_adultes * coef_adulte)
alimentation_crans_1 = alimentation_avec_loulou * coef_crans / (total_enfants * coef_enfant + total_adultes * coef_adulte)
alimentation_pau_1 = alimentation_avec_loulou * coef_pau / (total_enfants * coef_enfant + total_adultes * coef_adulte)
```

```python
assert alimentation_chalon + alimentation_olivet_1 + alimentation_crans_1 + alimentation_pau_1 == alimentation_avec_loulou
```

### Sans Loulou

```python
total_adultes = 5
total_enfants = 5  # Rose n'est pas comptabilisée
coef_adulte = 1
coef_enfant = 1
coef_olivet = 2 * coef_enfant + 2 * coef_adulte
coef_crans = 1 * coef_enfant + 2 * coef_adulte
coef_pau = 2 * coef_enfant + 1 * coef_adulte

assert coef_olivet + coef_crans + coef_pau == (total_enfants * coef_enfant + total_adultes * coef_adulte)
```

```python
alimentation_olivet_2 = alimentation_sans_loulou * coef_olivet / (total_enfants * coef_enfant + total_adultes * coef_adulte)
alimentation_crans_2 = alimentation_sans_loulou * coef_crans / (total_enfants * coef_enfant + total_adultes * coef_adulte)
alimentation_pau_2 = alimentation_sans_loulou * coef_pau / (total_enfants * coef_enfant + total_adultes * coef_adulte)

assert np.isclose(alimentation_olivet_2 + alimentation_crans_2 + alimentation_pau_2, alimentation_sans_loulou)
```

```python
print("Répartition des dépenses alimentaires :")
print(f"  Olivet : {alimentation_olivet_1 + alimentation_olivet_2:.2f} €")
print(f"  Crans : {alimentation_crans_1 + alimentation_crans_2:.2f} €")
print(f"  Pau : {alimentation_pau_1 + alimentation_pau_2:.2f} €")
print(f"  Chalon : {alimentation_chalon:.2f} €")
```

```python
assert np.isclose(alimentation_chalon + alimentation_olivet_1 + alimentation_olivet_2 + alimentation_crans_1 + alimentation_crans_2 + alimentation_pau_1 + alimentation_pau_2, alimentation_avec_loulou + alimentation_sans_loulou)
```
